# Working with Lambdas and Functional Programming Tools

````python
data = [(i-5) * (i-5) for i in range(10)]
print(data, " max: ", max(data), ", min: ", min(data))
iterable = range(10)  # Creates an iterable that goes from 0 to 9
print(iterable)
# dir(iterable)  # Shows what you can do with it.
some_list = list(iterable)  # Converts to a list. Python does not use proper 'arrays' naturally.
print(some_list)
some_set = set(iterable)
print(some_set)
some_dict = {i: i*i for i in iterable}  # This is the "List Comprehension" syntax
print(some_dict)
````

````python
# Lambdas - Functions can be treated as objects
print(lambda some_object: some_object)
x = lambda number: number + 1
y = x
print(y(1))
z = lambda number: x(x(number))
print(z(1))
print(list(map(z, iterable)))
````

````python
# filter and is_divisible
def is_divisible_by(factor):
    return (lambda number: number % factor == 0)

even = is_divisible_by(2)
print(list(filter(even, iterable)))
odd = lambda number: not(even(number))
print(list(filter(odd, iterable)))
````

````python
# map and multiply_by
def multiply_by(factor_to_multiply_by):
    return (lambda input_number: factor_to_multiply_by * input_number)

doubler = multiply_by(2)
print(list(map(doubler, some_list)))
tripler = multiply_by(3)
print(list(map(tripler, some_list)))
````

## The Radial Basis Function (RBF) kernel is one of the most commonly used kernels in Support Vector Machines (SVMs). Its particularly effective for handling non-linear classification problems by transforming the input space into a higher-dimensional feature space.

**Key Characteristics:**
1. **Non-linear Mapping**: The RBF kernel maps data points into a higher-dimensional space where a linear hyperplane can separate the data. This makes it suitable for non-linear problems.

*     Kernel Function: The RBF kernel is defined as: $$K(x, x') = \exp\left(-\gamma \|x - x'\|^2\right)$$
* Here:,: Two data points in the input space.






2
: The squared Euclidean distance between the two points.

* : A parameter that controls the influence of a single training example. Higher

 values focus more on closer neighbors, while lower values take distant points into account.

2. **Flexibility**: The RBF kernel adapts to data complexity and can model intricate decision boundaries without requiring an explicit feature map.

3. **Parameters**: SVM with an RBF kernel depends on two key parameters:

* C: Regularization parameter that balances the tradeoff between maximizing the
margin and minimizing classification error.
* 
: Influences the shape of the decision boundary.

**Advantages:**

* Effective for non-linear relationships.

* Requires fewer hyperparameters compared to some other kernels.

* Widely used due to its simplicity and power.

**Challenges:**

* Selection of appropriate 
 and

 values is critical and may require techniques like cross-validation.

* Computational cost can be high for large datasets due to pairwise computations.

````python
import math
import matplotlib.pyplot as plt
from sklearn import svm
import numpy as np

# Use increasing amounds of training data and measure performance on the same test set.
with open("./book_sales_work_data.json", "r") as infile:
    book_sales_work_data = json.load(infile)

train_data, test_data = model_selection.train_test_split(book_sales_work_data, test_size = .20)
print("Training Length: ", len(train_data), " Testing Length: ", len(test_data))

def convert_book_data_to_arrays(data_to_convert, regress=False):
    dtc = data_to_convert
    inputs = [[dtc[i]["book_details_time"], dtc[i]["homepage_time"],
               dtc[i]["total_web_time_minutes"]]
              for i in range(len(dtc))]
    if not regress:  # If for classification
        outputs = [dtc[i]["paying_customer"]
                  for i in range(len(dtc))]
    else:
        outputs = [dtc[i]["books_sold"]
          for i in range(len(dtc))]
    return inputs, outputs

# Split the test set into inputs and expected outputs
test_X, test_Y = convert_book_data_to_arrays(test_data)

num_parts = 100
data_sizes = [math.floor(1/num_parts * (i+1) * len(train_data)) for i in range(num_parts)] # measure by 10ths.
accuracies = []
for data_size in data_sizes:
    train_subset = train_data[:data_size]
    train_X, train_Y = convert_book_data_to_arrays(train_subset)

    # train model
    model = svm.SVC(kernel='rbf', gamma=0.7)
    model.fit(train_X,train_Y)

    # measure accuracy on the test data
    predicted_Y = model.predict(test_X)
    accuracy = np.mean([predicted_Y[i] == test_Y[i] for i in range(len(predicted_Y))])
    accuracies.append(accuracy)

# plot accuracy vs. number of training points.
plt.plot(data_sizes, accuracies)
plt.xlabel('Number of Training Points')
plt.ylabel('Test Accuracy')
plt.ylim(0,1)
plt.show()
````

### Boosting and Ensembles: A Comprehensive Overview

#### **Ensemble Methods**
Ensemble methods in machine learning involve combining the predictions of multiple models (base learners) to produce a more robust and accurate result than a single model. The idea is simple: by merging the strengths of several models, you reduce the risk of overfitting or errors caused by the limitations of individual models.

**Key Types of Ensemble Methods:**
1. **Bagging (Bootstrap Aggregating):**
   - Uses multiple models trained on different subsets of the data (created through bootstrapping).
   - Aggregates predictions through averaging (for regression) or majority voting (for classification).
   - Example: Random Forest.

2. **Boosting:**
   - Sequentially trains models, with each new model focusing on the errors of the previous ones.
   - Combines the predictions into a single, stronger model.
   - Example: AdaBoost, Gradient Boosting.

3. **Stacking:**
   - Combines predictions of multiple base models (called level-0 models) using a meta-model (called level-1 model) for the final prediction.

---

#### **Boosting**
Boosting is a specific type of ensemble technique designed to improve weak learners by sequentially building a strong learner. Unlike bagging, where models work independently, boosting trains models iteratively, correcting mistakes made by earlier models.

**How Boosting Works:**
1. Initially, a base model is trained on the dataset.
2. Misclassified samples (errors) are given higher weights so that subsequent models pay more attention to them.
3. New models are trained iteratively, focusing on minimizing the errors from the previous models.
4. Final predictions are a weighted sum (or combination) of the predictions from all the models.

---

#### **Key Boosting Algorithms**
1. **AdaBoost (Adaptive Boosting):**
   - Focuses on adjusting the weights of misclassified data points to guide subsequent learners.
   - Combines weak classifiers (e.g., decision stumps) into a single strong classifier.
   - Reduces bias and variance.

2. **Gradient Boosting:**
   - Optimizes a loss function by fitting new models to the residual errors (differences between actual and predicted values).
   - Often used for regression and classification problems.
   - Popular implementations include XGBoost, LightGBM, and CatBoost.

3. **Stochastic Gradient Boosting:**
   - Introduces randomness (e.g., subsets of data or features) during training to improve generalization and reduce overfitting.

---

#### **Advantages of Boosting**
- High accuracy on both structured and unstructured data.
- Effective in reducing bias and variance.
- Handles a wide variety of machine learning tasks, such as classification and regression.

#### **Challenges of Boosting**
- Prone to overfitting if not carefully tuned.
- Computationally intensive due to sequential training.
- Sensitive to noisy data and outliers.

---

**Summary:**
Ensemble methods like **boosting** leverage the power of multiple models to improve prediction accuracy. Boosting, in particular, focuses on learning from the mistakes of prior models and is widely used for high-performance tasks in machine learning.

### Gradient Boosting: A Deep Dive

Gradient Boosting is a powerful machine learning technique used for both regression and classification tasks. It builds an ensemble of weak learners (typically decision trees) in a sequential manner, where each new model corrects the errors of the previous ones. This iterative process optimizes a loss function to improve the overall predictive performance.

---

### Key Concepts of Gradient Boosting

1. **Weak Learners**:
   - Gradient Boosting typically uses shallow decision trees as weak learners.
   - These trees are simple models that perform slightly better than random guessing.

2. **Sequential Learning**:
   - Models are trained one after another, with each new model focusing on the residual errors (differences between actual and predicted values) of the previous model.

3. **Gradient Descent**:
   - The algorithm minimizes a specified loss function (e.g., mean squared error for regression or log-loss for classification) using gradient descent.
   - Each new model is trained to predict the negative gradient of the loss function with respect to the predictions.

4. **Additive Model**:
   - Predictions are made by summing the outputs of all the weak learners, weighted appropriately.

---

### Steps in Gradient Boosting

1. **Initialize the Model**:
   - Start with a simple model, such as predicting the mean of the target variable.

2. **Compute Residuals**:
   - Calculate the residuals (errors) between the actual values and the predictions.

3. **Fit a Weak Learner**:
   - Train a weak learner (e.g., a decision tree) to predict the residuals.

4. **Update Predictions**:
   - Add the predictions of the weak learner to the overall model, weighted by a learning rate.

5. **Repeat**:
   - Iterate the process, adding more weak learners to reduce the residual errors further.

---

### Key Parameters in Gradient Boosting

1. **Learning Rate (`eta`)**:
   - Controls the contribution of each weak learner to the final model.
   - Smaller values make the model more robust but require more iterations.

2. **Number of Trees (`n_estimators`)**:
   - The number of weak learners to train.
   - Too many trees can lead to overfitting, while too few may underfit.

3. **Tree Depth (`max_depth`)**:
   - Controls the complexity of each weak learner.
   - Deeper trees can capture more complex patterns but may overfit.

4. **Loss Function**:
   - Specifies the objective to minimize (e.g., mean squared error, log-loss).

5. **Subsampling (`subsample`)**:
   - Fraction of the training data used to train each tree.
   - Helps reduce overfitting and improves generalization.

---

### Advantages of Gradient Boosting

1. **High Accuracy**:
   - Gradient Boosting often outperforms other algorithms on structured data.

2. **Flexibility**:
   - Can handle various loss functions and be customized for specific tasks.

3. **Feature Importance**:
   - Provides insights into which features are most influential in predictions.

---

### Challenges of Gradient Boosting

1. **Overfitting**:
   - Can overfit if the number of trees or tree depth is too high.

2. **Computational Cost**:
   - Training can be slow, especially for large datasets.

3. **Hyperparameter Tuning**:
   - Requires careful tuning of parameters like learning rate and tree depth.

---

### Popular Implementations

1. **XGBoost**:
   - Highly efficient and scalable implementation of Gradient Boosting.
   - Includes regularization to reduce overfitting.

2. **LightGBM**:
   - Optimized for speed and memory efficiency.
   - Uses a histogram-based approach for faster training.

3. **CatBoost**:
   - Designed for categorical features.
   - Reduces the need for extensive preprocessing.

````python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F

# Define our neural network
class RegressionNet(nn.Module):
    def __init__(self, width):
        """width should be a natural number"""
        width = max(width, 1)

        super(self.__class__, self).__init__()

        #Define the pass-forward weights
        #Goes from 1 to width back to 1, with a nonlinear function in between.
        self.forward_1 = nn.Linear(3,width)
        self.forward_2 = nn.Linear(width,1)

    def forward(self, x):
        x = self.forward_2(F.relu(self.forward_1(x)))
        return x

    def infer_class(self, x):
        # Purely for convenience
        """Run the input forward, but then round to the nearest integer."""
        return int(math.floor(self.forward(x) + .5))

# Sometimes called the "criterion"
loss_function = nn.MSELoss()
````

````python
def variable_wrap(some_data, convert_type=np.float64):
    return Variable(torch.from_numpy(np.array(some_data).astype(convert_type)).float())

def list_map(*args, **kwargs):
    # A helper function to unroll a list generated by mapping a function
    #     onto elements of another list.
    # args stores unnamed arguments, and kwargs stores named arguments
    return list(map(*args,**kwargs))
````

````python
from torch.optim import Adam

def learn_on_data(train_X, train_Y, test_X, test_Y, regress=False, model_width=10, file_base="./"):
    model = RegressionNet(model_width)
    optimizer = Adam(model.parameters())

    # Format the input to be read by the network.
    var_train_X, var_train_Y = list_map(variable_wrap, train_X), list_map(variable_wrap, train_Y)
    var_test_X, var_test_Y = list_map(variable_wrap, test_X), list_map(variable_wrap, test_Y)

    train_accuracies = []
    test_accuracies = []

    for data_loop in range(300):
        model.train() # Set the network to training mode. Affects some layers sometimes.

        optimizer.zero_grad()
        # Loop through the data and use the gradient to optimize the model parameters.
        for train_index in range(len(train_X)):
            X, Y = var_train_X[train_index], var_train_Y[train_index]

            #print(Y)
            predicted_Y = model.forward(X)
            #print(predicted_Y.item())
            loss = loss_function(predicted_Y, Y)
            #print(loss.item())
            loss.backward() # Stores the gradient for all parameters.
        optimizer.step() # Doing this outside the loop accumulates gradients across all data points.

        # Measure train and test accuracy
        model.eval() # Set model to evaluation (testing) mode. Affects some models some time.
        if not regress:
            train_predicted = list_map(model.infer_class, var_train_X)
        else:
            train_predicted = list_map(model.forward, var_train_X)

        if not regress:
            matches = [train_predicted[i]==var_train_Y[i] for i in range(len(var_train_X))]
            matches = list_map(int, matches) # Convert to Boolean to higher-represention-size int
            train_accuracy = sum(matches)/len(var_train_X)
        else:
            # In this setting measure error as absolute value of the difference
            error = [abs(train_predicted[i].detach()-var_train_Y[i]) for i in range(len(var_train_X))]
            #error = [(train_predicted[i]-var_train_Y[i])**2 for i in range(len(var_train_X))]
            average_error = sum(error)/len(var_train_X)
            train_accuracy = average_error
        train_accuracies.append(train_accuracy)

        if not regress:
            test_predicted = list_map(model.infer_class, var_test_X)
        else:
            test_predicted = list_map(model.forward, var_test_X)

        if not regress:
            matches = [test_predicted[i]==var_test_Y[i] for i in range(len(var_test_X))]
            matches = list_map(int, matches)
            test_accuracy = sum(matches)/len(var_test_X)
        else:
            error = [abs(test_predicted[i].detach()-var_test_Y[i]) for i in range(len(var_test_X))]
            #error = [(test_predicted[i]-var_test_Y[i])**2 for i in range(len(var_test_X))]
            average_error = sum(error)/len(var_test_X)
            test_accuracy = average_error
        test_accuracies.append(test_accuracy)

    if not regress:
        plt.plot(range(len(train_accuracies)), train_accuracies, label="training_accuracy")
        plt.plot(range(len(test_accuracies)), test_accuracies, label="testing_accuracy")
        plt.xlabel('Passes through data')
        plt.ylabel('Accuracy (higher is better)')
        plt.title('Classification with a Neural Network')
    else:
        plt.plot(range(len(train_accuracies)), train_accuracies, label="training_accuracy")
        plt.plot(range(len(test_accuracies)), test_accuracies, label="testing_accuracy")
        plt.xlabel('Passes through data')
        plt.ylabel('Average Error (lower is better)')
        plt.title('Regression with a Neural Network')
    plt.legend()
    plt.show()

learn_on_data(class_train_X, class_train_Y, class_test_X, class_test_Y, regress=False)
learn_on_data(regress_train_X, regress_train_Y, regress_test_X, regress_test_Y, regress=True)
````

K-Means Clustering

**k-Means** is an unsupervised machine learning algorithm used for **clustering**, where data points are grouped into a predefined number of clusters (**k**). The algorithm aims to minimize the variance within each cluster and maximize the variance between clusters by iteratively assigning data points to their nearest cluster centroid and updating the centroid positions based on the mean of the assigned points.

---

### Key Steps of k-Means:
1. **Initialize Centroids**: Randomly select `k` points as initial cluster centroids.
2. **Assign Points**: Assign each data point to the cluster with the nearest centroid (using Euclidean distance or another metric).
3. **Update Centroids**: Calculate the mean of all points in each cluster and update the centroid to this mean.
4. **Repeat**: Iterate steps 2 and 3 until the centroids stabilize (no significant change) or a set number of iterations is reached.

---

### Key Features:
- **Input**: Number of clusters `k` needs to be defined beforehand.
- **Output**: Cluster assignments for each data point.
- **Applications**: Customer segmentation, image compression, anomaly detection, and more.
- **Limitations**: Sensitive to the choice of `k`, initial centroids, and outliers.

### Clustering and Bayesian Inference - Expectation Maximization

**Bayesian inference** in clustering involves using Bayes' theorem to model the uncertainty in assigning data points to clusters, incorporating prior beliefs about the data and updating these beliefs as new information becomes available. In this approach, clustering is treated as a probabilistic problem where clusters are considered latent variables.

---

### Key Points:
1. **Probabilistic Framework**:
   - Each data point is assumed to belong to a cluster with a certain probability.
   - Bayesian inference calculates the posterior probability of a data point belonging to a cluster using prior knowledge (priors) and observed data (likelihood).

2. **Model-Based Clustering**:
   - Bayesian methods often involve a mixture model, like a **Gaussian Mixture Model (GMM)**, where each cluster is represented by a probability distribution.
   - Parameters of these distributions (e.g., mean and variance) are estimated iteratively using algorithms like **Expectation-Maximization (EM)**.

3. **Flexibility**:
   - Unlike k-means, Bayesian clustering can automatically determine the number of clusters by treating the number of clusters as a random variable (e.g., using Dirichlet Process Mixture Models).

4. **Applications**:
   - Bayesian clustering is commonly used in areas like genetics (e.g., clustering gene expression data) and topic modeling (e.g., Latent Dirichlet Allocation).

This method accounts for uncertainty and provides a richer, probabilistic understanding of the clustering process compared to deterministic methods like k-means. Let me know if you'd like an example or more details! 

**Expectation-Maximization (EM)** is an iterative optimization algorithm used in statistics and machine learning to estimate the parameters of probabilistic models, especially when the data contains hidden or latent variables. It is widely applied in scenarios like clustering, mixture models (e.g., Gaussian Mixture Models), and missing data problems.

---

### Key Idea:
The EM algorithm alternates between two steps:
1. **Expectation (E-Step):**
   - Computes the expected value of the latent variables given the observed data and current estimates of the model parameters. This step involves calculating a **posterior distribution** over the latent variables.

2. **Maximization (M-Step):**
   - Updates the model parameters to maximize the likelihood of the observed data, incorporating the expectations computed in the E-step.

The algorithm repeats these two steps iteratively until it converges to a local maximum of the likelihood function.

---

### How It Works (Mathematical Outline):
1. Let \( X \) be the observed data, \( Z \) the latent variables, and \( \theta \) the model parameters.
2. The goal is to maximize the **log-likelihood** of the observed data:
   \[
   \log P(X | \theta)
   \]
3. Since \( Z \) is unobserved, the algorithm maximizes the **expected log-likelihood**:
   \[
   Q(\theta | \theta^{(t)}) = \mathbb{E}_{Z}[ \log P(X, Z | \theta) | X, \theta^{(t)} ]
   \]
   - \( \theta^{(t)} \) represents the parameters from the previous iteration.

4. **Steps**:
   - **E-Step**: Compute \( Q(\theta | \theta^{(t)}) \), the expectation of the complete log-likelihood given the current parameter estimates.
   - **M-Step**: Maximize \( Q(\theta | \theta^{(t)}) \) with respect to \( \theta \) to update the parameters.

---

### Applications:
1. **Gaussian Mixture Models (GMMs):**
   - EM is used to estimate the means, variances, and mixing coefficients of the Gaussian distributions in a GMM.
   
2. **Clustering:**
   - When using probabilistic models (e.g., soft clustering), EM assigns probabilities of belonging to clusters instead of hard labels.
   
3. **Hidden Markov Models (HMMs):**
   - EM (in the form of the Baum-Welch algorithm) is used to estimate the transition and emission probabilities in HMMs.

4. **Missing Data:**
   - EM handles datasets with missing values by iteratively estimating the missing data and updating model parameters.

---

### Strengths:
- Handles hidden variables and incomplete data effectively.
- Flexible and broadly applicable to various probabilistic models.

### Weaknesses:
- Sensitive to initialization; it can converge to a local, rather than global, maximum.
- Computationally expensive for large datasets or complex models.

In summary, EM is a cornerstone algorithm for parameter estimation in probabilistic models, providing a structured way to deal with uncertainty and latent variables. Let me know if youd like me to dive into a specific application, such as GMMs or HMMs!

**Hierarchical clustering** is a method of clustering data points into a hierarchy of nested clusters. Unlike flat clustering methods like k-means, hierarchical clustering builds a multi-level structure that allows exploration of clusters at different levels of granularity. It works well when you want to understand the structure of your data or visualize relationships in a dendrogram.

---

### Types of Hierarchical Clustering:

1. **Agglomerative Hierarchical Clustering (Bottom-Up)**:
   - Starts with each data point as its own individual cluster.
   - Iteratively merges the closest clusters until all data points belong to one single cluster.
   - Most commonly used.

2. **Divisive Hierarchical Clustering (Top-Down)**:
   - Starts with all data points in one cluster.
   - Iteratively splits clusters into smaller clusters until each data point is its own cluster.
   - Less commonly used due to computational complexity.

---

### Key Steps in Hierarchical Clustering:

1. **Compute a Distance Matrix**:
   - Measure the distance between all pairs of data points using metrics like:
     - **Euclidean distance**
     - **Manhattan distance**
     - **Cosine similarity**

2. **Linkage Criteria**:
   - Determines how the distance between clusters is calculated. Examples include:
     - **Single Linkage**: Distance between the closest points of two clusters.
     - **Complete Linkage**: Distance between the farthest points of two clusters.
     - **Average Linkage**: Average distance between all points in two clusters.
     - **Centroid Linkage**: Distance between the centroids of two clusters.

3. **Build the Dendrogram**:
   - The clustering process is visualized using a dendrogram, a tree-like diagram that shows the order and distance at which clusters are merged or split.

4. **Cut the Dendrogram**:
   - Determine the number of clusters by cutting the dendrogram at a certain level.

---

### Pros of Hierarchical Clustering:
- Does not require the number of clusters to be pre-specified (unlike k-means).
- Produces a hierarchy, allowing exploration of clusters at different levels.
- Useful for small to medium-sized datasets.

### Cons of Hierarchical Clustering:
- Computationally intensive for large datasets due to the need to compute distances between all pairs.
- Sensitive to noise and outliers, which can distort the clustering structure.

---

### Applications:
- **Gene Expression Analysis**: Clustering genes or samples with similar behavior.
- **Document Clustering**: Organizing similar documents into nested categories.
- **Market Segmentation**: Grouping customers based on purchasing behavior.

**Data whitening**, also known as **sphering**, is a preprocessing technique used to transform data to make it uncorrelated and normalize its variance. The purpose of whitening is to remove redundancy and make the features of a dataset have the same scale, which is often beneficial for machine learning algorithms, especially those sensitive to covariance or feature scaling.

---

### Key Characteristics:
1. **Removes Correlation**:
   - Whitening eliminates linear correlations between features by transforming the covariance matrix of the data into an identity matrix (diagonal matrix where all off-diagonal elements are zero).
   
2. **Normalizes Variance**:
   - All features are scaled to have unit variance (variance = 1).

---

### Steps in Data Whitening:
1. **Mean-Center the Data**:
   - Subtract the mean of each feature to center the data around zero.
   - \( X_{\text{centered}} = X - \text{mean}(X) \)

2. **Compute the Covariance Matrix**:
   - Calculate the covariance matrix of the centered data.

3. **Perform Eigenvalue Decomposition**:
   - Decompose the covariance matrix into its eigenvectors and eigenvalues.

4. **Transform the Data**:
   - Scale the data by the square root of the eigenvalues (to normalize variance) and rotate the data using the eigenvectors to decorrelate the features.

---

### Whitening Formula:
Let \( X \) be the centered data, \( \Sigma \) the covariance matrix, \( E \) the eigenvectors, and \( \Lambda \) the eigenvalues of \( \Sigma \):
\[
X_{\text{whitened}} = \Lambda^{-\frac{1}{2}} E^T X
\]

Alternatively, **Principal Component Analysis (PCA)** can also be used for whitening by reducing dimensionality first, then scaling.

---

### Applications:
- **Machine Learning**:
   - Improves convergence of gradient-based optimizers by standardizing the data.
   - Often used before applying algorithms like PCA, ICA, or neural networks.
   
- **Signal Processing**:
   - Removes noise and redundancy in signal data.
   
- **Computer Vision**:
   - Normalizes image data for feature extraction and classification.

---

### Drawbacks:
- **Loss of Interpretability**: Transformed features may no longer have a direct physical interpretation.
- **Numerical Stability**: Whitening can amplify noise if the original data has very small eigenvalues.

**Principal Component Analysis (PCA)** is a dimensionality reduction technique widely used in machine learning and data analysis. It transforms high-dimensional data into a lower-dimensional space while preserving as much variance (information) as possible.

---

### Key Features:
1. **Dimensionality Reduction**: PCA reduces the number of features by finding new axes (called principal components) that maximize the variance in the data.
2. **Orthogonal Components**: Principal components are uncorrelated (orthogonal) to one another.
3. **Applications**: Used for data visualization, noise reduction, and speeding up machine learning algorithms.

---

### How It Works:
- **Step 1**: Standardize the data (mean = 0, variance = 1).
- **Step 2**: Compute the covariance matrix.
- **Step 3**: Perform eigenvalue decomposition on the covariance matrix to find eigenvectors (principal components) and eigenvalues (variance explained by each component).
- **Step 4**: Select the top components to transform the data into a lower-dimensional space.

---

# Product Recommendations Using Semi-Supervised Learning

Define semi-supervised learning
Cluster to transfer recipe for label transduction

The **cluster-to-transfer recipe for label transduction** is a method used in machine learning and data analysis to assign labels to unlabeled data points by leveraging clustering techniques. This approach is particularly useful in semi-supervised learning, where only a subset of the data is labeled.

---

### Key Steps in the Cluster-to-Transfer Recipe:
1. **Clustering**:
   - The algorithm first groups the data into clusters based on similarity (e.g., using k-means, hierarchical clustering, or Gaussian Mixture Models).
   - Each cluster represents a group of data points that are likely to share similar characteristics.

2. **Label Transfer**:
   - Labels from the labeled data points are propagated to the unlabeled points within the same cluster.
   - This assumes that data points in the same cluster are likely to belong to the same class.

3. **Refinement**:
   - The process may include iterative refinement to improve the accuracy of label assignments, often by re-clustering or adjusting cluster boundaries.

---

### Applications:
- **Single-Cell Data Analysis**:
   - Used to transfer cell type labels from a reference dataset to a new dataset in biological studies.
- **Image and Text Classification**:
   - Helps in labeling large datasets where manual annotation is impractical.
- **Semi-Supervised Learning**:
   - Bridges the gap between labeled and unlabeled data for better model training.

This method is particularly effective when the clusters are well-separated and the labeled data is representative of the overall dataset. Let me know if you'd like to explore specific examples or tools that implement this approach! 

**Label Propagation** is a semi-supervised learning algorithm that is used to propagate labels from labeled data points to unlabeled data points within a connected graph structure. The main idea is that similar or connected points are likely to share the same label. This approach is particularly useful when only a small portion of the dataset is labeled, and the rest needs to be classified.

---

### Key Concepts of Label Propagation:
1. **Graph Representation**:
   - The algorithm represents the dataset as a graph, where:
     - Nodes represent data points.
     - Edges represent similarities or distances between the points (e.g., using a similarity matrix).

2. **Propagation of Labels**:
   - Initially, labels are assigned to labeled nodes, while unlabeled nodes are left blank.
   - In each iteration, the labels from labeled nodes are "propagated" through the graph to unlabeled nodes based on edge weights (similarities).

3. **Convergence**:
   - The algorithm continues propagating labels until a convergence criterion is met (e.g., labels no longer change significantly).

---

### Steps in Label Propagation:
1. **Build a Similarity Graph**:
   - Compute a similarity matrix (e.g., Gaussian Kernel or k-nearest neighbors) to construct edges between nodes based on feature similarity.

2. **Initialize Labels**:
   - Assign known labels to labeled points and set unknown labels to a default state.

3. **Iterative Propagation**:
   - At each step, update the label of each node as the weighted average (or majority vote) of the labels of its neighbors.

4. **Final Label Assignment**:
   - Once the algorithm converges, assign the most probable label to each unlabeled node.

---

### Applications:
- **Social Networks**: Label nodes (e.g., users or groups) based on partial information, such as friend connections.
- **Image Segmentation**: Assign labels (e.g., object types) to pixels or regions in an image.
- **Document Classification**: Classify documents based on similarities and a small set of labeled examples.
- **Protein Interaction Networks**: Predict unknown interactions by propagating known labels.

---

### Benefits:
- Efficient and straightforward to implement.
- No need to set a predefined number of clusters (as in k-means).
- Exploits both labeled and unlabeled data, making it ideal for semi-supervised tasks.

### Limitations:
- Sensitive to the choice of similarity measure or graph structure.
- Propagation errors in labels can spread to other nodes, leading to inaccuracies.

**Nearest Neighbor Search (NNS)** is a fundamental technique in data science and machine learning used to efficiently find the data point(s) in a dataset that are closest to a given query point. It is widely applied in tasks like recommendation systems, clustering, computer vision, and natural language processing.

---

### **What is Nearest Neighbor Search?**
Given:
- A dataset of points in a multi-dimensional space.
- A query point \( q \).

The goal of NNS is to find the point(s) in the dataset that minimizes the distance to \( q \). The "distance" is typically calculated using a metric such as:
- **Euclidean distance** (common for continuous data).
- **Manhattan distance**.
- **Cosine similarity** (for high-dimensional data like text embeddings).
- **Hamming distance** (for binary data).

---

### **Key Variants of NNS:**
1. **Exact Nearest Neighbor Search**:
   - Finds the actual nearest neighbor(s) with complete accuracy.
   - Computationally expensive for large datasets as it typically requires scanning all points (\( O(n) \)).

2. **Approximate Nearest Neighbor Search (ANN)**:
   - A faster method that approximates the nearest neighbors with high probability of correctness.
   - Useful for very large datasets or high-dimensional spaces.

---

### **Algorithms for Nearest Neighbor Search:**
1. **Brute-Force Search**:
   - Computes the distance from the query point to all points in the dataset.
   - Accurate but inefficient for large datasets or high-dimensional data (\( O(n) \)).

2. **Tree-Based Methods**:
   - Partition the space to reduce the search space.
   - **k-d Tree**:
     - Divides the space along the data dimensions for efficient querying.
     - Works well in lower dimensions (\(<20\)).
   - **Ball Tree**:
     - Groups data into hyperspheres, improving performance in moderately high dimensions.

3. **Approximate Methods**:
   - Speed up the search process at the cost of some accuracy.
   - **Locality-Sensitive Hashing (LSH)**:
     - Hashes similar points into the same bucket to reduce the search space.
     - Works well for high-dimensional data.
   - **Hierarchical Navigable Small World (HNSW)**:
     - Constructs a graph for efficient multi-layer neighbor navigation.

---

### **Applications of Nearest Neighbor Search:**
1. **Recommendation Systems**:
   - Finding similar users/items (e.g., recommending movies based on user preferences).

2. **Clustering**:
   - k-Means clustering requires finding the closest cluster centroid for each point.

3. **Image Retrieval**:
   - Search for visually similar images by comparing feature embeddings.

4. **Text Similarity**:
   - Identify documents or sentences similar to a query text using cosine similarity in embedding space.

5. **Real-Time Systems**:
   - Autonomous vehicles use NNS for pathfinding and obstacle avoidance.

````python
def vector_similarities(v1, v2):
    # Compute number of matching fields (-1 does not match by default)
    agreement = sum([v1[i] == v2[i] for i in range(len(v1)) if v1[i] > -1 ])
    return agreement

def confidence(target, neighbor):
    # Many items do they have in common?
    return sum([1 for i in range(len(neighbor)) if neighbor[i] > -1 and target[i] > -1])

def fill_field(index, target, data, k):
    fill_data = [elem for elem in data if elem[index] > -1] # if the field is filled
    k = min(k, len(fill_data))
    similarities = [vector_similarities(target, data_elem) for data_elem in fill_data]
    confidences = [confidence(target, data_elem) for data_elem in fill_data]
    # take the k most similar items
    similarity_cutoff = list(reversed(sorted(similarities)))[k-1]
    sum_weights = sum([confidences[i] for i in range(len(fill_data)) if similarities[i]>=similarity_cutoff])
    sum_votes = sum([fill_data[i][index] * confidences[i] for i in range(len(fill_data)) if similarities[i]>=similarity_cutoff])
    return sum_votes/sum_weights, k

def fill_vector(query_vector, data, k):
    scores = []
    confidences = []
    for index in range(len(query_vector)):
        if query_vector[index] > -1:
            scores.append(query_vector[index])
            confidences.append(-1) # Meaning infinite confidence
        else:
            score, confidence = fill_field(index, query_vector, data, k)
            scores.append(score)
            confidences.append(confidence)
    return scores, confidences
````

## Thompson Sampling
Step1. Model each action with a mean reward and uncertainty
Step2. Sample  potential reward from each action based on the distribution
Step3. Take the action with the highest sampled potential reward

Using Thompson Sampling for:
* Testing page layouts
* Testing advertisements
* Optimizing notifications
Optimizing marketing funnels
* testing new service models

Benefits
> Fewer customer get the bad options
we can test way more than 2 things
Tests can be run for much longer
Because data efficiency is huge.

Best case scenarion: Frequent decisions, near-instant feedback - lots of possibilities

````python
import numpy as np
import math

def mean_std(list_of_numbers):
    return np.mean(list_of_numbers), np.std(list_of_numbers)

def mean_ste(list_of_numbers):
    mean, std = mean_std(list_of_numbers)
    return mean, std/math.sqrt(len(list_of_numbers))

def biased(list_of_numbers):
    #Prevent std from being zero by adding phantom observations.
    new_list = list_of_numbers.copy()
    new_list.extend([1, 0])
    return new_list

class GaussianThompsonSampler:
    """Gaussians describe the statistics around means very well.
        This class implements Thompson Sampling based on this."""

    @staticmethod
    def sample_arms(means, std_errs):
        """Takes arm statistics and returns a sampled index.
            std_err is deviation in the mean, called the standard error."""
        # means and std_devs are assumed to correspond element for element.
        samples = means.copy()
        for index in range(len(means)):
            if std_errs[index] > 0:
                samples[index] = np.random.normal(means[index], std_errs[index])
        max_ind = max(range(len(means)), key=lambda index: samples[index])
        return max_ind

    @staticmethod
    def sample_from_scores(scores_matrix):
        """scores_matrix is a list of lists of scores."""
        means = []
        std_errs = []
        for scores_list in scores_matrix:
            mean, ste = mean_ste(biased(scores_list))
            means.append(mean)
            std_errs.append(ste)
        return GaussianThompsonSampler.sample_arms(means, std_errs)
````

````python
class InfiniteThompsonSampler:
    @staticmethod
    def sample_arms(means, std_errs):
        """Takes arm statistics and returns a sampled index.
            std_err is deviation in the mean, called the standard error."""
        # means and std_devs are assumed to correspond element for element.
        samples = means.copy()
        for index in range(len(means)):
            if std_errs[index] > 0:
                samples[index] = np.random.normal(means[index], std_errs[index])
        # Define the mean and std_dev of value for a new arm
        new_mean, new_std = mean_std(biased(means)) # Prevent deterministic non-sampling
        samples.append(np.random.normal(new_mean, new_std))
        max_ind = max(range(len(samples)), key=lambda index: samples[index])
        return max_ind

    @staticmethod
    def sample_from_scores(scores_matrix):
        """scores_matrix is a list of lists of scores."""
        means = []
        std_errs = []
        for scores_list in scores_matrix:
            mean, ste = mean_ste(biased(scores_list))
            means.append(mean)
            std_errs.append(ste)
        return InfiniteThompsonSampler.sample_arms(means, std_errs)
````
